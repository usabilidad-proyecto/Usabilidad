```python
from pandas.io.parsers import read_csv
import matplotlib.pyplot as plt
import numpy as np
import glob

%matplotlib inline

fileToRead = open("Profiler.txt","r")
projectName = fileToRead.read().splitlines()

def lee_csv(folder):
    valores = []
    if (folder == "Editor"):
        path = "../Assets/ProfilerLogs/*.csv"
    else:
        path = projectName[0] + "_Data/ProfilerLogs/" + folder +"/*.csv"
    for fname in glob.glob(path):
        valores.append(read_csv(fname, header=None, delimiter=';').values)
    return valores

valoresHW = lee_csv("HW")
valoresUnity = lee_csv("Unity")
valoresEditor = lee_csv("Editor")

for v in valoresHW:
    valoresGHW = v[1:].T

    plt.xlabel('Segundos de ejecucin')
    plt.ylabel('%')
    plt.plot(valoresGHW[0] ,valoresGHW[1], color = 'red', marker = '.', label = "Process RAM")
    plt.legend()
   #plt.plot(valoresGHW[0],valoresGHW[3], color = 'orange', marker = '.')
    plt.show()
    plt.plot(valoresGHW[0],valoresGHW[2], color = 'green', marker = '.', label = "Process CPU")
    plt.plot(valoresGHW[0],valoresGHW[4], color = 'blue', marker = '.', label = "Total CPU Usage")
    plt.legend()
    plt.show()
for v in valoresUnity:
    valoresGUnity = v.T

    plt.xlabel('Segundos de ejecucin')
    plt.ylabel('int')

    plt.plot(valoresGUnity[1] ,valoresGUnity[2], color = 'red', marker = '.', label = "FPS")
    plt.legend()
    plt.show()
    plt.plot(valoresGUnity[1] ,valoresGUnity[3], color = 'yellow', marker = '.', label = "Objects")
    plt.legend()
    plt.show()
    
for v in valoresEditor:
    valoresGEditor = v.T

    plt.xlabel('Segundos de ejecucin')
    plt.ylabel('int')

    plt.plot(valoresGEditor[1] ,valoresGEditor[3], color = 'blue', marker = '.', label = "DrawCalls")
    plt.legend()
    plt.gca().axes.get_xaxis().set_visible(False)
    plt.show()
    plt.plot(valoresGEditor[1] ,valoresGEditor[4], color = 'orange', marker = '.', label = "Triangles")
    plt.legend()
    plt.gca().axes.get_xaxis().set_visible(False)
    plt.plot(valoresGEditor[1] ,valoresGEditor[5], color = 'black', marker = '.', label = "Vertex")
    plt.legend()
    plt.gca().axes.get_xaxis().set_visible(False)
    plt.show()
    plt.gca().axes.get_xaxis().set_visible(False)
    plt.plot(valoresGEditor[1] ,valoresGEditor[6], color = 'green', marker = '.', label = "Objects")
    plt.legend()
    plt.gca().axes.get_xaxis().set_visible(False)
    plt.show()
    plt.plot(valoresGEditor[1] ,valoresGEditor[7], color = 'green', marker = '.', label = "RenderCalls")
    plt.legend()
    plt.gca().axes.get_xaxis().set_visible(False)
    plt.show()
```
